```python


import heapq

# --------- BASE DE CONOCIMIENTO ----------
transmilenio = {
    "Portal Norte": {"Cardio Infantil": 2},
    "Cardio Infantil": {"Portal Norte": 2, "Calle 146": 1},
    "Calle 146": {"Cardio Infantil": 1, "Calle 142": 1},
    "Calle 142": {"Calle 146": 1, "Calle 140": 1},
    "Calle 140": {"Calle 142": 1, "Calle 127": 2},
    "Calle 127": {"Calle 140": 2, "Calle 106": 2},
    "Calle 106": {"Calle 127": 2, "Calle 100": 1},
    "Calle 100": {"Calle 106": 1, "Virrey": 1},
    "Virrey": {"Calle 100": 1, "Calle 85": 1},
    "Calle 85": {"Virrey": 1, "Calle 76": 1},
    "Calle 76": {"Calle 85": 1, "Calle 72": 1},
    "Calle 72": {"Calle 76": 1, "Calle 63": 2},
    "Calle 63": {"Calle 72": 2, "Calle 57": 1},
    "Calle 57": {"Calle 63": 1, "Calle 45": 2},
    "Calle 45": {"Calle 57": 2, "Calle 34": 2},
    "Calle 34": {"Calle 45": 2, "Calle 26": 2},
    "Calle 26": {"Calle 34": 2, "Centro": 1},
    "Centro": {"Calle 26": 1, "Hospital": 2},
    "Hospital": {"Centro": 2, "Portal Sur": 3},
    "Portal Sur": {"Hospital": 3}
}

# --------- HEURÍSTICA ----------
def heuristica(a, b):
    return 0   # heurística simple para no complicarlo


# --------- ALGORITMO A* ----------
def a_star(inicio, destino):
    cola = []
    heapq.heappush(cola, (0, inicio))

    costos = {inicio: 0}
    padres = {inicio: None}

    while cola:
        costo_actual, actual = heapq.heappop(cola)

        if actual == destino:
            break

        for vecino, peso in transmilenio[actual].items():
            nuevo_costo = costo_actual + peso

            if vecino not in costos or nuevo_costo < costos[vecino]:
                costos[vecino] = nuevo_costo
                prioridad = nuevo_costo + heuristica(vecino, destino)
                heapq.heappush(cola, (prioridad, vecino))
                padres[vecino] = actual

    # reconstruir la ruta
    ruta = []
    nodo = destino
    while nodo is not None:
        ruta.append(nodo)
        nodo = padres[nodo]
    ruta.reverse()

    return ruta, costos[destino]


# --------- INTERFAZ ---------
print("=== Sistema de rutas en TransMilenio ===")
print("Estaciones disponibles:")
for est in transmilenio.keys():
    print(" -", est)

inicio = input("\nIngrese estación de inicio: ")
destino = input("Ingrese estación de destino: ")

if inicio not in transmilenio or destino not in transmilenio:
    print("\n❌ Error: revise los nombres.")
else:
    ruta, costo = a_star(inicio, destino)

    print("\nRuta recomendada:\n")
    for r in ruta:
        print("  →", r)

    print("\nDistancia aproximada:", costo)
    print("(Estos valores son ficticios, solo para el ejercicio.)")
```
